```python
y = df_new.total_cases.values
```

```python
X = df_new.drop(['week_start_date', 'total_cases'], axis = 1).values
```

```python
test = df.drop(['week_start_date'], axis = 1).values
```

Splitting into training and validation data

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.1, random_state=1, shuffle = True)
```

```python
len(X_train)
```

```python
len(X_val)
```

```python
len(y_train)
```

```python
missing_val = IterativeImputer(random_state=1)
```

Imputing missing values

```python
X_train = missing_val.fit_transform(X_train)
X_val = missing_val.transform(X_val)
test = missing_val.transform(test)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
normal_scale = StandardScaler().fit(X_train)
```

```python
X_train = normal_scale.transform(X_train)
```

```python
X_val = normal_scale.transform(X_val)
```

```python
test = normal_scale.transform(test)
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
model= AdaBoostRegressor(base_estimator=GradientBoostingRegressor(max_depth=5, min_samples_split=2, loss='ls',n_estimators=400, learning_rate=.1, random_state=1)).fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
y_predict = model.predict(X_val)
y_pred = y_predict.astype(int)
```

```python
print(model.score(X_val, y_val)), print(r2_score(y_pred, model.predict(X_val)))
```

```python
df_val=pd.DataFrame({'Actual': y_val, 'Predicted':y_pred})
df_val
```
